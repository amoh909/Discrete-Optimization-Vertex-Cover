```python
import networkx as nx
import sys; sys.path.insert(0, '../src')
from lp_solver import solve_vertex_cover_lp

G1 = nx.Graph()
G1.add_edge(0, 1)
G2 = nx.path_graph(3)
G3 = nx.cycle_graph(3)
G4 = nx.cycle_graph(4)

result1 = solve_vertex_cover_lp(G1)
result2 = solve_vertex_cover_lp(G2)
result3 = solve_vertex_cover_lp(G3)
result4 = solve_vertex_cover_lp(G4)

print("Status:", result1["status"])
print("Objective:", result1["objective"])
print("x values:", result1["x_values"])

print("Status:", result2["status"])
print("Objective:", result2["objective"])
print("x values:", result2["x_values"])

print("Status:", result3["status"])
print("Objective:", result3["objective"])
print("x values:", result3["x_values"])

print("Status:", result4["status"])
print("Objective:", result4["objective"])
print("x values:", result4["x_values"])

def check_lp_constraints(G, x_values):
    for u, v in G.edges():
        if x_values[u] + x_values[v] < 1 - 1e-9:
            return False
    return True

print("Constraints satisfied:", check_lp_constraints(G1, result1["x_values"]))
print("Constraints satisfied:", check_lp_constraints(G2, result2["x_values"]))
print("Constraints satisfied:", check_lp_constraints(G3, result3["x_values"]))
print("Constraints satisfied:", check_lp_constraints(G4, result4["x_values"]))
```

```python
import networkx as nx
import sys; sys.path.insert(0, '../src')
from ilp_solver import solve_vertex_cover_ilp

def test_graph(G, name):
    result = solve_vertex_cover_ilp(G)

    print(f"--- {name} ---")
    print("Status:", result["status"])
    print("Objective:", result["objective"])
    print("x values:", result["x_values"])
    print("cover:", result["cover"])
    print("Cover size matches objective:",
      len(result["cover"]) == int(round(result["objective"])))


# 1. Single edge
G1 = nx.Graph()
G1.add_edge(0, 1)

# 2. Path graph with 3 nodes
G2 = nx.path_graph(3)

# 3. Triangle
G3 = nx.cycle_graph(3)

# 4. 4-cycle
G4 = nx.cycle_graph(4)

test_graph(G1, "Single Edge")
test_graph(G2, "Path Graph (3 nodes)")
test_graph(G3, "Triangle")
test_graph(G4, "4-Cycle")

```
